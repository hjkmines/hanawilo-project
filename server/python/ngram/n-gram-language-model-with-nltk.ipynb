```python
#from nltk.lm.preprocessing import pad_both_ends
#from nltk.lm.preprocessing import flatten
import pandas as pd 
from nltk.tokenize import word_tokenize
import nltk
from nltk.lm.preprocessing import padded_everygram_pipeline
```

Get the csv file where data were scraped from Udemy webpages.

```python
table = pd.read_csv('../output/udemy_courses_python_2023-01-11.csv')
```

```python
table.drop(['Unnamed: 0'],axis=1)
```

```python
text = table['Course Title']
```

```python
text=text.tolist()
```

Get 1542 course titles as a list

```python
text
```

```python
""" import string
text=text.translate(str.maketrans('', '', string.punctuation)) """
```

```python
# Tokenize and lower first letter of the text.
tokens=[word_tokenize(x.lower()) for x in text]
```

```python
tokens
```

```python
# Preprocess the tokenized text for 3-grams language modelling
n = 3
train_data, padded_sents = padded_everygram_pipeline(n, tokens)
```

```python
from nltk.lm import MLE
model = MLE(n) # Lets train a 3-grams model, previously we set n=3
```

Initializing the MLE model, creates an empty vocabulary

```python
len(model.vocab)
```

... which gets filled as we fit the model.

```python
model.fit(train_data, padded_sents)
print(model.vocab)
```

```python
model.score('language')
```

```python
len(model.vocab)
```

```python
print(model.vocab.lookup(tokens[0])) # the first index of tokens
```

```python
print(model.counts) # There are 51456 combinations of 3 ngram orders.
```

Example of a maximum likelihood.

```python
model.counts[['100']]['days'] # i.e. Count('100','days')
```

```python
model.counts[['100','days']]['of'] # i.e. Count('100', 'days','of')
```

```python
model.score('of', '100 days'.split())  # P('of'|'100','days')
```

Generate a sequence of words by given one or several words and a fixed length.

```python
print(model.generate(10, text_seed=['python'],random_seed=7)) # Generate a 10 word length sentence with given first word python.
```

Remove end sentence tags of the generated sentence.

```python
from nltk.tokenize.treebank import TreebankWordDetokenizer

detokenize = TreebankWordDetokenizer().detokenize

def generate_sent(model, num_words,text_seed,random_seed):
    """
    :param model: An ngram language model from `nltk.lm.model`.
    :param num_words: Max no. of words to generate.
    :param random_seed: Seed value for random.
    """
    content = text_seed
    for token in model.generate(num_words, text_seed = text_seed,random_seed=random_seed):
        if token == '<s>':
            continue
        if token == '</s>':
            break
        content.append(token)
    return detokenize(content)
```

```python
generate_sent(model, 10, ['python'],7)
```
